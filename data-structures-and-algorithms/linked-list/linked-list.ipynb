```python
%%file l.c
#include <stdio.h>
#include <stdlib.h>

struct Node {

    int data;
    struct Node* next;
};

struct Node* head = NULL;

void create(int A[], int n){

    head = (struct Node*)malloc(sizeof(struct Node));
    head->data = A[0];
    head->next = NULL;
    struct Node* current = head;

    for(int i=1;i<n;i++){
        struct Node* t = (struct Node*)malloc(sizeof(struct Node));
        current->next = t;
        t->data = A[i];
        t->next = NULL;
        current = t;
    }
    
}

void display(struct Node* node){

    // Time O(n)
    // Space O(1)

    struct Node* ptr = node;
    
    
    while(ptr != NULL){
        printf("%d ",ptr->data);
        ptr = ptr->next;
    }
    
    
}

void rDisplay(struct Node* node){

    // Reccurrence Relation : T(n) = T(n) + 1
    // Time O(n)
    // space O(n)
    
    if(node != NULL){
        rDisplay(node->next);
        printf("%d ",node->data);
    }
}


int main() {

    int A[] = {1,2,3,4,5,6};
    int n = 6;

    create(A,n);

    display(head);
    printf("\n");
    rDisplay(head);

    return 0;
}
```
