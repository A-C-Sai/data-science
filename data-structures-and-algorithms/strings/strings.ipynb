#include <stdio.h>
int main() {
    
    // reversing string 
    char A[7] = "pyhton";
    char B[7];

    int i = 0;
    while(A[i+1] != '\0') i++;

    int j= 0;
    while(i>=0) B[j++] = A[i--];
    B[j] = '\0';

    printf("%s",B);
    return 0;
}

```C
#include <stdio.h>
int main() {

    // reversing string
    char A[7] = "pyhton";

    printf("%s\n",A);

    int i = 0;
    while(A[i+1] != '\0') i++;

    int j= 0;
    while(j<i) {
        char t = A[i];
        A[i] = A[j];
        A[j] = t;
        j++;i--;
    }
    

    printf("%s\n",A);
    return 0;
}
```

```C
#include <stdio.h>
int main() {

    // string comparison (case sensitive)

    char A[] = "Paint";
    char B[] = "Painting";

    int i,j;

    for(i=0,j=0; A[i] && B[i] ;i++,j++){
        if(A[i] != B[j]) break;
    }

    if(A[i]==B[j]) printf("%d\n",0); // equal
    else if(A[i] > B[j]) printf("%d\n",1); // greater
    else printf("%d\n",-1); // smaller


    return 0;
}
```

    // Palindrome
    char A[] = "malayalam";

    int i;
    for(i=0;A[i+1]!='\0';i++);

    int j;
    for(j=0;j<i;j++,i--){
        if(A[j]!=A[i]){
            printf("%s\n","Not a Palindrome");

        }
    }
    printf("%s\n","Palindrome");

}
