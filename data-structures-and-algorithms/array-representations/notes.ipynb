```python
'''
Array - collection of elements; all elements are of same type. Similar data elements grouped under one name.
        Memory allocated will be contiguous.


    e.g. int A[] = {1,2,3,4,5};

         printf("%d",A[2]); // 3
         printf("%d",2[A]); // 3
         printf("%d",*(A+2)); // 3

Once an array is created it's size cannot be modified (size decided at complie time; memory allocated at run-time)

Dynamic Arrays can be created using Heap memory
Heap memory can be accessed indirectly with the help of pointers

Note: Same Array size cannot be grown as memory allocated must be contiguous. Instead another Array of desired
      size is created and elements are copied to the new array.
'''
print()
```

```python
%%file array.c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>
int main() {

    int n = 5;
    int* p = (int*)malloc(n*sizeof(int));

    for(int i=0;i<n;i++){
        p[i] = i+1;
    }

    for(int j=0;j<n;j++){
        printf("%d ",p[j]);
    }
    printf("\n");

    int m = 10;

    /*
    int* q = (int*)malloc(m*sizeof(int));
    memcpy(q,p,n*sizeof(p[0]));

    free(p);
    p=q;
    q=NULL;
    */
    


    p = (int*)realloc(p,m*sizeof(int));

    for(int k=0;k<m;k++){
        printf("%d ",p[k]);
    }
    printf("\n");

   

    free(p);
    p=NULL;

    printf("p is pointing to = %p\n",p);

    
    

    return 0;
}
```

```python
%%bash 
gcc array.c -o array
./array
```

```python
%%file array1.c
#include <stdio.h>
#include <stdlib.h>
int main() {

    // 2-D Array in heap

    int rows = 3;
    int cols = 4;

    int* A[rows]; // array of pointers in stack

    for(int i=0;i<rows;i++){
        A[i] = (int*)malloc(cols*sizeof(int));
    }

    for(int r=0;r<rows;r++){
        for(int c=0;c<cols;c++) {
            *(*(A+r)+c) = r+c;
        }
    }

    for(int r=0;r<rows;r++){
        for(int c=0;c<cols;c++) {
            printf("%d\t",*(*(A+r)+c));
        }
        printf("\n");
    }

    for(int i=0;i<rows;i++){
        free(A[i]);
    }

    return 0;

}
```

```python
%%bash
gcc array1.c -o a1
./a1
```

```python
%%file array2.c
#include <stdio.h>
#include <stdlib.h>
int main() {

    int rows = 3;
    int cols = 4;

    int* *ptr = (int**)malloc(rows*sizeof(int*));
    
    for(int r=0;r<rows;r++){
        *(ptr+r) = (int*)malloc(cols*sizeof(int));
    }

    for(int r=0;r<rows;r++) {
        for(int c=0;c<cols;c++) {
            *(*(ptr+r)+c) = r+c;
        }
    }

    for(int r=0;r<rows;r++) {
        for(int c=0;c<cols;c++) {
            printf("%d\t",ptr[r][c]);
        }
        printf("\n");
    }

    for(int r=0;r<rows;r++){
        free(*(ptr+r));
    }

    free(ptr);
    ptr=NULL;
    

    return 0;
}
```

```python
%%bash
gcc array2.c -o a2
./a2
```
